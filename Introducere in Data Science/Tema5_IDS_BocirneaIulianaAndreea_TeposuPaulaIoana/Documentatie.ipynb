# Linear Regresion

Regresia liniară este o metodă folosită pentru predicţia unei valori numerice dintr-o mulţime infinită de valori. 

În regresia liniară, relațiile sunt modelate după o funcție predictivă liniară ,iar parametrii sunt estimați din datele de intrare. Astfel de modele se numesc modele liniare.

Regresia liniară a fost primul tip de analiză de regresie, care a fost studiată riguros și care a fost utilizată pe scară largă în aplicații practice. Acest lucru se datorează faptului că modelele care depind liniar de parametrii lor necunoscuți sunt mai ușor de adaptat decât modelele care sunt neliniar legate de parametrii lor.

Modelele de regresie liniară sunt de obicei modelate folosind metoda least squares.De asemenea, această metodă poate fi folosită și asupra modelelor neliniare.

Pe scurt, acest algoritm trasează o dreaptă care să aproximeze cât mai bine datele din setul de intrare. Se dorește ca eroarea dată de estimare să fie cât mai mică. Eroarea este influențată de diferența dintre valoarea prezisă și valoarea actuală.




În imagine este reprezentată cu roșu dreapta determinată de algoritm, punctele albastre reprezintă valorile actuale din setul de date, iar liniile verzi reprezintă eroare dată de estimare. 

În biblioteca sklearn există clasa LinearRegression care ne ajută să facem astfel de estimări.

Această clasă are ca parametrii:<br>
<b>-fit_intercept</b> care este default=True și sugerează că se realizează calculul de interceptare asupra modelului. Dacă aceste parametru este False, se presupune că datele sunt centrate.<br>
<b>-normalize</b> este default=False și este ignorat dacă fit_intercept este False. Dacă este setat la True atunci se realizează normalizarea setului de date.<br>
<b>-copy_X </b> este default=True și arată dacă se face o copie a setului de intrare sau se suprascrie.<br>
<b>-n_jobs </b>reprezintă numărul job-urilor<br>
<b>-positive </b>care este default=False , iar dacă este setat la True forțează coeficienții să devină pozitivi.<br>

# DecisionTreeRegressor

Arborii de decizie pot fi folosiți atât în probleme de clasificare cât și în probleme de regresie.<br>
Un arbore decizional este un model de predicție care folosește un set de "binary rules"  pentru a calcula o anumită valoare.
Fiecare arbore conține ramuri, noduri și frunze.<br> Arborele decizional se construiește pe baza unor întrebări. Fiecare întrebare împarte setul de date în subseturi până se ajunge la frunze.
Întrebările au formă de tipul True/False. Pentru fiecare răspuns de tipul True sau False apar alte ramuri în arbore.<br>
Ca orice model de învățare supervizată, este nevoie de o fază de antrenare. Predicția pe care o va da va fi o estimare bazată pe setul de date pe care a fost antrenat.<br>Regresia cu arbori decizionali folosește MSE pentru a decide dacă un nod este împărțit în alte subnoduri. Acești arbori sunt folosiți de obicei pentru a modela curbe sinusoidale cu zgomot adăugat.

În imagine, datele sunt reprezentate sub forma unei curbe sinusoidale. Algoritmul determină niște regresii liniare locale care sunt aproximate pe curbă. Se observă că dacă parametrul max_depth are o valoare mare, modelul învață foarte multe detalii fine din setul de andrenare cât și din zgomotul adăugat.<br>
În biblioteca sklearn există clasa DecisionTreeRegressor care folosește principiile prezentate mai sus.<br> Parametrii săi sunt:<br>
<b>-criterion</b> care reprezintă criteriul după care se face separarea nodurilor în subnoduri.Valoarea acestul parametru este default=mse<br>
<b>-splitter</b> reprezintă strategia de împărțire a nodurilor. Poate lua valoare best sau random<br>
<b>-max_depth</b> înălțimea maximă a arborelui<br>

# Logistic Regresion

În statistică, modelele logistice sunt folosite pentru a modela probabilitatea unei anumite clase sau a unui anumit eveniment precum admis/respind, câștig/pierdere, viu/mort etc. Poate fi extins pentru a modela mai multe clase sau mai multe evenimente,de exemplu identificarea unui anumit animal dintr-o imagine.<br>
Regresia logistică este o tehnică fundamentală de clasificare. Aparține grupului de clasificatori lineari și este oarecum similar cu regresia polinomială și liniară. Regresia logistică este rapidă și relativ simplă și este convenabil pentru  interpretarea rezultatelor. Deși este în esență o metodă de clasificare binară, poate fi aplicată și problemelor cu mai multe clase.<br>
În analiza regresiei, regresia logistică estimează parametrii unui model logistic. Matematic, un model logistic binar are o variabilă dependentă cu două valori posibile, cum ar fi trecere / eșuare care este reprezentată printr-o variabilă indicator, unde cele două valori sunt etichetate „0” și „1”.<br>
Regresia logistică este reprezentată sub forma unei ecuații, asemănător cu regresia liniară. Valorile de intrare sunt combinate în funcție de coeficienți pentru a prezice output-ul. Diferența față de regeresia liniară este că output-ul este o valoare binară, ci nu una numerică.
Coeficienții pentru regresia logistică sunt estimați pe baza setului de antrenare. Cei mai buni coeficienți ar rezulta într-un model care ar prezice o valoare foarte apropiată de 1 pentru clasa implicită și o valoare foarte apropiată de 0 pentru cealaltă clasă.


Pentru realizarea regersiei logistice în python, există clasa LogisticRegression cu următorii parametrii:<br>
<b>-penalty</b> care reprezintă norma folosită pentru penalizare.<br>
<b>-fit_intercept</b> care specifică dacă o constantă trebuie adăugată la funcția de decizie.Valoarea sa default este True.<br>
<b>-solver</b> algoritmul folosit pentru optimizarea problemei<br>
<b>-max_iter</b> numărul maxim de iterații<br>

# ARD (Automatical Relevance Determination) Regression

Regresia ARD folosește modelul Bayesian Ridge Regression.<br>
BayesianRidge estimează un model probabilistic al problemei de regresie.Rezultatul acestui model se numește Bayesian Ridge Regression.<br>
În comparație cu estimatorul OLS, coeficienții sunt shift-ați către 0 pentru a se stabiliza.
Estimarea modelului se face prin maximizarea iterativă log-likelihood asupra datelor.


Clasa ARDRegression din biblioteca sklearn adaptează modelul de regresie folosind un ARD prior. Se presupune că modelul de regresie se află în distribuție Gaussiană. De asemenea, este estimat parametrul lambda care reprezintă precizia distribuției datelor  și alfa care reprezintă precizia distribuției zgomotului.
Estimarea se face printr-o procedură iterativă(Evidence Maximization).<br>
Parametrii pentru ARDRegression sunt:<br>
<b>-n_iter</b> numărul maxim de iterații<br>
<b>-fit_intercept</b> care este default=True și sugerează că se realizează calculul de interceptare asupra modelului. Dacă aceste parametru este False, se presupune că datele sunt centrate.<br>
<b>-normalize</b> este default=False și este ignorat dacă fit_intercept este False. Dacă este setat la True atunci se realizează normalizarea setului de date.<br>
<b>-copy_X </b> este default=True și arată dacă se face o copie a setului de intrare sau se suprascrie.<br>

# Epsilon-SupportVectorRegression

Support Vector Machine (SVM) este o unealtă populară în problemele de clasificare și de regresie. În regresia Epsilon-SVR setul de antrenare conține variabile predictive și valori de răspuns observate.
SVR ne dă posibilitatea de a alege cât de mari să fie valorile acceptate pentru erorile din modelul nostru.
Obiectivul funcției SVR este de a minimiza coeficienții.<br>

În imaginea de mai sus este reprezentat un set de antrenare. Linia roșie reprezintă valorile modelate corect, iar linia neagră reprezintă marja de eroare.Se poate observa că algoritmul nu funcționează pentru toate datele din setul de antrenare. Trebuie să ținem cont și de valorile care se află în afara lui epsilon.Pentru fiecare valoare din afara lui epsilon , reținem abaterea sa de la marjă.Deoarece aceste abateri pot exista, ele se adaugă la funcția inițială.<br>
Această abatere determină apariția unui nou parametru C care crește sau scade toleranța asupara valorilor din afara lui epsilon.<br>
SVR este un algoritm puternic care ne permite să alegem cât de toleranți suntem față de erori, atât printr-o marjă de eroare acceptabilă (ϵ), cât și prin reglarea toleranței noastre de a cădea în afara ratei de eroare acceptată.<br>
Clasa Epsilon-SupportVectorRegression este implementată in biblioteca sklearn și are următorii paramatrii:<br>

<b>-kernel</b> specifică tipul de kernel care este folosit pentru aplicarea algoritmului. Valoarea sa default este rbf.<br>
<b>-gamma</b> coeficientul pentru kernel-ul de tip rbf ,poly și signoid.Poate lua valorile auto sau scale.<br>
<b>-epsilon</b> coeficientul epsilon<br>
<b>-C</b> parametrul de regularizare<br>
