# Linear Regresion

Regresia liniar este o metod folosit pentru predicia unei valori numerice dintr-o mulime infinit de valori. 

n regresia liniar, relaiile sunt modelate dup o funcie predictiv liniar ,iar parametrii sunt estimai din datele de intrare. Astfel de modele se numesc modele liniare.

Regresia liniar a fost primul tip de analiz de regresie, care a fost studiat riguros i care a fost utilizat pe scar larg n aplicaii practice. Acest lucru se datoreaz faptului c modelele care depind liniar de parametrii lor necunoscui sunt mai uor de adaptat dect modelele care sunt neliniar legate de parametrii lor.

Modelele de regresie liniar sunt de obicei modelate folosind metoda least squares.De asemenea, aceast metod poate fi folosit i asupra modelelor neliniare.

Pe scurt, acest algoritm traseaz o dreapt care s aproximeze ct mai bine datele din setul de intrare. Se dorete ca eroarea dat de estimare s fie ct mai mic. Eroarea este influenat de diferena dintre valoarea prezis i valoarea actual.




n imagine este reprezentat cu rou dreapta determinat de algoritm, punctele albastre reprezint valorile actuale din setul de date, iar liniile verzi reprezint eroare dat de estimare. 

n biblioteca sklearn exist clasa LinearRegression care ne ajut s facem astfel de estimri.

Aceast clas are ca parametrii:<br>
<b>-fit_intercept</b> care este default=True i sugereaz c se realizeaz calculul de interceptare asupra modelului. Dac aceste parametru este False, se presupune c datele sunt centrate.<br>
<b>-normalize</b> este default=False i este ignorat dac fit_intercept este False. Dac este setat la True atunci se realizeaz normalizarea setului de date.<br>
<b>-copy_X </b> este default=True i arat dac se face o copie a setului de intrare sau se suprascrie.<br>
<b>-n_jobs </b>reprezint numrul job-urilor<br>
<b>-positive </b>care este default=False , iar dac este setat la True foreaz coeficienii s devin pozitivi.<br>

# DecisionTreeRegressor

Arborii de decizie pot fi folosii att n probleme de clasificare ct i n probleme de regresie.<br>
Un arbore decizional este un model de predicie care folosete un set de "binary rules"  pentru a calcula o anumit valoare.
Fiecare arbore conine ramuri, noduri i frunze.<br> Arborele decizional se construiete pe baza unor ntrebri. Fiecare ntrebare mparte setul de date n subseturi pn se ajunge la frunze.
ntrebrile au form de tipul True/False. Pentru fiecare rspuns de tipul True sau False apar alte ramuri n arbore.<br>
Ca orice model de nvare supervizat, este nevoie de o faz de antrenare. Predicia pe care o va da va fi o estimare bazat pe setul de date pe care a fost antrenat.<br>Regresia cu arbori decizionali folosete MSE pentru a decide dac un nod este mprit n alte subnoduri. Aceti arbori sunt folosii de obicei pentru a modela curbe sinusoidale cu zgomot adugat.

n imagine, datele sunt reprezentate sub forma unei curbe sinusoidale. Algoritmul determin nite regresii liniare locale care sunt aproximate pe curb. Se observ c dac parametrul max_depth are o valoare mare, modelul nva foarte multe detalii fine din setul de andrenare ct i din zgomotul adugat.<br>
n biblioteca sklearn exist clasa DecisionTreeRegressor care folosete principiile prezentate mai sus.<br> Parametrii si sunt:<br>
<b>-criterion</b> care reprezint criteriul dup care se face separarea nodurilor n subnoduri.Valoarea acestul parametru este default=mse<br>
<b>-splitter</b> reprezint strategia de mprire a nodurilor. Poate lua valoare best sau random<br>
<b>-max_depth</b> nlimea maxim a arborelui<br>

# Logistic Regresion

n statistic, modelele logistice sunt folosite pentru a modela probabilitatea unei anumite clase sau a unui anumit eveniment precum admis/respind, ctig/pierdere, viu/mort etc. Poate fi extins pentru a modela mai multe clase sau mai multe evenimente,de exemplu identificarea unui anumit animal dintr-o imagine.<br>
Regresia logistic este o tehnic fundamental de clasificare. Aparine grupului de clasificatori lineari i este oarecum similar cu regresia polinomial i liniar. Regresia logistic este rapid i relativ simpl i este convenabil pentru  interpretarea rezultatelor. Dei este n esen o metod de clasificare binar, poate fi aplicat i problemelor cu mai multe clase.<br>
n analiza regresiei, regresia logistic estimeaz parametrii unui model logistic. Matematic, un model logistic binar are o variabil dependent cu dou valori posibile, cum ar fi trecere / euare care este reprezentat printr-o variabil indicator, unde cele dou valori sunt etichetate 0 i 1.<br>
Regresia logistic este reprezentat sub forma unei ecuaii, asemntor cu regresia liniar. Valorile de intrare sunt combinate n funcie de coeficieni pentru a prezice output-ul. Diferena fa de regeresia liniar este c output-ul este o valoare binar, ci nu una numeric.
Coeficienii pentru regresia logistic sunt estimai pe baza setului de antrenare. Cei mai buni coeficieni ar rezulta ntr-un model care ar prezice o valoare foarte apropiat de 1 pentru clasa implicit i o valoare foarte apropiat de 0 pentru cealalt clas.


Pentru realizarea regersiei logistice n python, exist clasa LogisticRegression cu urmtorii parametrii:<br>
<b>-penalty</b> care reprezint norma folosit pentru penalizare.<br>
<b>-fit_intercept</b> care specific dac o constant trebuie adugat la funcia de decizie.Valoarea sa default este True.<br>
<b>-solver</b> algoritmul folosit pentru optimizarea problemei<br>
<b>-max_iter</b> numrul maxim de iteraii<br>

# ARD (Automatical Relevance Determination) Regression

Regresia ARD folosete modelul Bayesian Ridge Regression.<br>
BayesianRidge estimeaz un model probabilistic al problemei de regresie.Rezultatul acestui model se numete Bayesian Ridge Regression.<br>
n comparaie cu estimatorul OLS, coeficienii sunt shift-ai ctre 0 pentru a se stabiliza.
Estimarea modelului se face prin maximizarea iterativ log-likelihood asupra datelor.


Clasa ARDRegression din biblioteca sklearn adapteaz modelul de regresie folosind un ARD prior. Se presupune c modelul de regresie se afl n distribuie Gaussian. De asemenea, este estimat parametrul lambda care reprezint precizia distribuiei datelor  i alfa care reprezint precizia distribuiei zgomotului.
Estimarea se face printr-o procedur iterativ(Evidence Maximization).<br>
Parametrii pentru ARDRegression sunt:<br>
<b>-n_iter</b> numrul maxim de iteraii<br>
<b>-fit_intercept</b> care este default=True i sugereaz c se realizeaz calculul de interceptare asupra modelului. Dac aceste parametru este False, se presupune c datele sunt centrate.<br>
<b>-normalize</b> este default=False i este ignorat dac fit_intercept este False. Dac este setat la True atunci se realizeaz normalizarea setului de date.<br>
<b>-copy_X </b> este default=True i arat dac se face o copie a setului de intrare sau se suprascrie.<br>

# Epsilon-SupportVectorRegression

Support Vector Machine (SVM) este o unealt popular n problemele de clasificare i de regresie. n regresia Epsilon-SVR setul de antrenare conine variabile predictive i valori de rspuns observate.
SVR ne d posibilitatea de a alege ct de mari s fie valorile acceptate pentru erorile din modelul nostru.
Obiectivul funciei SVR este de a minimiza coeficienii.<br>

n imaginea de mai sus este reprezentat un set de antrenare. Linia roie reprezint valorile modelate corect, iar linia neagr reprezint marja de eroare.Se poate observa c algoritmul nu funcioneaz pentru toate datele din setul de antrenare. Trebuie s inem cont i de valorile care se afl n afara lui epsilon.Pentru fiecare valoare din afara lui epsilon , reinem abaterea sa de la marj.Deoarece aceste abateri pot exista, ele se adaug la funcia iniial.<br>
Aceast abatere determin apariia unui nou parametru C care crete sau scade tolerana asupara valorilor din afara lui epsilon.<br>
SVR este un algoritm puternic care ne permite s alegem ct de tolerani suntem fa de erori, att printr-o marj de eroare acceptabil (), ct i prin reglarea toleranei noastre de a cdea n afara ratei de eroare acceptat.<br>
Clasa Epsilon-SupportVectorRegression este implementat in biblioteca sklearn i are urmtorii paramatrii:<br>

<b>-kernel</b> specific tipul de kernel care este folosit pentru aplicarea algoritmului. Valoarea sa default este rbf.<br>
<b>-gamma</b> coeficientul pentru kernel-ul de tip rbf ,poly i signoid.Poate lua valorile auto sau scale.<br>
<b>-epsilon</b> coeficientul epsilon<br>
<b>-C</b> parametrul de regularizare<br>
